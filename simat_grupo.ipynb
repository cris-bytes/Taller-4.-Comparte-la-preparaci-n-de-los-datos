# Análisis de Clustering K-Means para la Población Estudiantil en SIMAT Cali

Este análisis aplica el algoritmo K-Means para agrupar estudiantes basados en características clave, con el fin de identificar perfiles que informen la política pública relacionada con el Trastorno del Espectro Autista (TEA) y otras necesidades de apoyo en Cali.

## Selección de Características y Preprocesamiento

Para el clustering, se seleccionaron las siguientes variables:

*   **Edad:** Edad del estudiante.
*   **ESTRATO:** Estrato socioeconómico.
*   **COMUNA:** Comuna de residencia.
*   **GRADO:** Grado de escolaridad.

Estas variables fueron elegidas por su relevancia potencial para identificar diferentes perfiles de estudiantes y posibles factores de riesgo o necesidades de apoyo.

Se realizó un preprocesamiento para preparar los datos para K-Means:

1.  Verificación y manejo de valores faltantes.
2.  Codificación One-Hot para la variable categórica 'COMUNA'.
3.  Escalado de las variables numéricas ('Edad', 'ESTRATO', 'GRADO') para asegurar que tengan una escala similar y no dominen el cálculo de distancias en el algoritmo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Configurar el estilo de los gráficos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12

# Cargar el dataset (asumiendo que 'df' ya está cargado en el entorno)
# Si no está cargado, descomenta y ejecuta la siguiente línea:
# df = pd.read_excel('/content/POBLACIÖN TEA CALI.xlsx')

# --- Selección de Características ---
selected_features_final = ['Edad', 'ESTRATO', 'COMUNA', 'GRADO']
df_clustering_final = df[selected_features_final].copy()

# --- Preprocesamiento ---

# Verificar y manejar valores faltantes (si los hay en las columnas seleccionadas)
print("Valores faltantes en las características seleccionadas antes del preprocesamiento:")
display(df_clustering_final.isnull().sum())

# Codificación One-Hot para 'COMUNA'
df_clustering_final = pd.get_dummies(df_clustering_final, columns=['COMUNA'], drop_first=True)

# Identificar columnas numéricas para escalar
numerical_cols_final = ['Edad', 'ESTRATO', 'GRADO']
categorical_cols_encoded_final = [col for col in df_clustering_final.columns if col.startswith('COMUNA_')]

# Separar columnas numéricas y codificadas
df_numerical_final = df_clustering_final[numerical_cols_final]
df_categorical_encoded_final = df_clustering_final[categorical_cols_encoded_final]

# Escalar columnas numéricas
scaler_final = StandardScaler()
df_numerical_scaled_final = scaler_final.fit_transform(df_numerical_final)
df_numerical_scaled_final = pd.DataFrame(df_numerical_scaled_final, columns=numerical_cols_final)

# Combinar columnas escaladas y codificadas
df_clustering_processed_final = pd.concat([df_numerical_scaled_final, df_categorical_encoded_final], axis=1)

print("\nPrimeras filas del DataFrame preprocesado y listo para clustering:")
display(df_clustering_processed_final.head())
print("\nInformación del DataFrame preprocesado:")
display(df_clustering_processed_final.info())
```

## Determinación del Número Óptimo de Clusters (k)

Se aplica el Método del Codo para visualizar la reducción de la inercia (suma de cuadrados dentro del cluster) a medida que aumenta el número de clusters. El "codo" en el gráfico sugiere un número óptimo de clusters donde la reducción de la inercia comienza a disminuir significativamente.

```python
# --- Determinación del Número Óptimo de Clusters (k) ---
inertia_final = []
k_range_final = range(1, 11) # Rango para probar k

for k in k_range_final:
    kmeans_elbow_final = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_elbow_final.fit(df_clustering_processed_final)
    inertia_final.append(kmeans_elbow_final.inertia_)

# Graficar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(k_range_final, inertia_final, marker='o')
plt.title("Método del Codo para Determinar el Número Óptimo de Clusters")
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Inercia")
plt.xticks(k_range_final)
plt.grid(True)
plt.show()
```

## Implementación y Entrenamiento del Modelo K-Means

Basado en la observación del gráfico del Método del Codo, se selecciona un número óptimo de clusters. Luego se entrena el modelo K-Means final con este número y se asignan las etiquetas de cluster a cada estudiante en el DataFrame original.

```python
# --- Implementación y Entrenamiento de K-Means ---
# ***AJUSTA ESTE VALOR BASADO EN TU INTERPRETACIÓN DEL GRÁFICO DEL CODO ANTERIOR***
optimal_k_final = 3 # Número óptimo de clusters (ejemplo: ajusta según el gráfico)
# ***************************************************************************

kmeans_final_model = KMeans(n_clusters=optimal_k_final, random_state=42, n_init=10)
kmeans_final_model.fit(df_clustering_processed_final)

# Asignar las etiquetas de los clusters al DataFrame original
df['Cluster_Final'] = kmeans_final_model.labels_

print(f"Modelo K-Means entrenado con {optimal_k_final} clusters.")
display(df[['Edad', 'ESTRATO', 'COMUNA', 'GRADO', 'Cluster_Final']].head())
```
